**Data Visualization with Matplotlib and Seaborn**

**Data Visualization :**
- Data visualization is the graphical representation of information and data. 
- It helps in identifying patterns, trends, and outliers in data using visual elements like charts, graphs, and maps.

 **1.Theory: Learn about Matplotlib (basic plots) and Seaborn (advanced plots like heatmaps, pairplots).**

**Matplotlib (Basic Plots):**
- Matplotlib is a foundational Python library used for creating static, animated, and interactive plots. 
- Think of it as the basic plotting tool very flexible but can be verbose.
- The Basic Plots are

**a) Line Plot**
- Used to visualize trends over time or continuous data.

```python
import matplotlib.pyplot as plt
x = [1, 2, 3, 4, 5]
y = [2, 3, 7, 1, 0]
plt.plot(x, y)
plt.title("Line Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

**b) Bar Chart**
- Represents categorical data with rectangular bars.

```python
categories = ['A', 'B', 'C']
values = [10, 24, 36]
plt.bar(categories, values)
plt.title("Bar Chart")
plt.show()
```

**c) Histogram**
- It Shows distribution of a single numerical variable.
- X-axis: Continuous intervals (bins) of the data.
- Y-axis: Frequency (count) of values within each bin.
- Bars: Each bar represents the frequency of data within a particular range.

```python
data = [22, 25, 27, 29, 30, 32, 35, 35, 36, 40, 41, 42, 45, 50, 52, 55]
plt.hist(data, bins=5, edgecolor='black')
plt.title('Histogram')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

**d) Scatter Plot**
- Displays relationship between two numeric variables.

```python
x = [1, 2, 3, 4, 5]
y = [2, 4, 1, 8, 7]
plt.scatter(x, y)
plt.title("Scatter Plot")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

**e) Pie Chart**
- Shows proportions in a circular format.

```python
labels = ['A', 'B', 'C']
sizes = [30, 45, 25]
plt.pie(sizes, labels=labels, autopct='%1.1f%%')
plt.title("Pie Chart")
plt.show()
```

**Seaborn (advanced plots like heatmaps, pairplots).**
- Seaborn is built on top of Matplotlib and provides a high-level interface for attractive and informative statistical graphics.

**a) Heatmap**
- Shows a matrix-style plot of values using color intensity.

```python
import seaborn as sns
import numpy as np
data = np.random.rand(5, 5)
sns.heatmap(data, annot=True, cmap='coolwarm')
plt.title("Heatmap")
plt.show()
```

**b) Pairplot**
- A pairplot is a powerful visualization tool used in exploratory data analysis (EDA) to show pairwise relationships between numerical variables in a dataset.
- It’s especially useful for identifying patterns, correlations, or clusters.
- A grid of scatter plots for each pair of features.
- The diagonal often shows a distribution (histogram or KDE) of each variable.
- Used commonly with Pandas DataFrames and built using Seaborn

```python
import seaborn as sns
import matplotlib.pyplot as plt
from seaborn import load_dataset
# Example using built-in dataset
df = sns.load_dataset('Iris')  # flower dataset with species
sns.pairplot(df, hue='species')  # hue adds color by species category
plt.show()
```

**c) Boxplot**
- Visualizes distribution, median, quartiles, and outliers.
-  A box plot (also called a box-and-whisker plot) is a graphical representation used to display the distribution, spread, and outliers of a dataset. 

- Minimum (excluding outliers)
- First Quartile (Q1) – 25th percentile
- Median (Q2) – 50th percentile
- Third Quartile (Q3) – 75th percentile
- Maximum (excluding outliers)
- Outliers – Points that fall outside 1.5 × IQR (interquartile range)

```python
sns.boxplot(x='species', y='sepal_length', data=df)
plt.title("Boxplot")
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv(r"C:\Users\prati\OneDrive\Datafiles\Iris.csv")
# Set plot style
sns.set(style="whitegrid")
# List of numerical features
features = ['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']
# Generate box plots
for feature in features:
    plt.figure(figsize=(8, 5))
    sns.boxplot(x='Species', y=feature, data=df)
    plt.title(f'{feature} Distribution by Species')
    plt.xlabel('Species')
    plt.ylabel(feature)
    plt.tight_layout()
    plt.show()
```

**d) Violin Plot**
- They combine a box plot (in the center) with a kernel density estimation (KDE) on each side.

- These plots provide a richer view of the distribution shape (e.g., bimodal, skewed).

- Vertical thickness indicates data density: the thicker the plot, the more data points fall in that range.

```python
# Generate violin plots
for feature in features:
    plt.figure(figsize=(8, 5))
    sns.violinplot(x='Species', y=feature, data=df, inner='box')  # 'inner=box' overlays a box plot
    plt.title(f'{feature} Distribution by Species (Violin Plot)')
    plt.xlabel('Species')
    plt.ylabel(feature)
    plt.tight_layout()
    plt.show()
```

**e) Count Plot**
- It shows count of observations in each category.

```python
# Set the visual style
sns.set(style="ticks")
# Create the count plot
plt.figure(figsize=(6, 4))
sns.countplot(x='Species', data=df)
plt.title('Count of Each Iris Species')
plt.xlabel('Species')
plt.ylabel('Count')
plt.tight_layout()
plt.show()
```

**Visualize the Dataset**
- Count Plot (Check Class Balance)

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.countplot(x='Species', data=df)
plt.title('Count of Each Iris Species')
plt.show()
```

**Box Plots (Compare Distributions by Species)**

```python
features = ['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']
for feature in features:
    sns.boxplot(x='Species', y=feature, data=df)
    plt.title(f'{feature} Distribution by Species')
    plt.show()
```

**Scatter Plot**

```python
# Scatter plot: PetalLength vs PetalWidth
colors = {'Iris-setosa':'red', 'Iris-versicolor':'green', 'Iris-virginica':'blue'}
plt.figure(figsize=(7, 5))
for species in df['Species'].unique():
    subset = df[df['Species'] == species]
    plt.scatter(subset['PetalLengthCm'], subset['PetalWidthCm'],
                label=species, color=colors[species])
plt.xlabel('Petal Length (cm)')
plt.ylabel('Petal Width (cm)')
plt.title('Petal Length vs Petal Width')
plt.legend()
plt.grid(True)
plt.show()
```

 **Violin Plots (Show Distribution + Density)**

```python
for feature in features:
    sns.violinplot(x='Species', y=feature, data=df)
    plt.title(f'{feature} Distribution by Species (Violin Plot)')
    plt.show()
```

**Pairplot (See Feature Relationships)**

```python
sns.pairplot(df, hue='Species')
plt.suptitle('Pairwise Feature Relationships', y=1.02)
plt.show()
```

**Correlation Heatmap**

```python
from sklearn.preprocessing import LabelEncoder
df_encoded = df.copy()
for col in df_encoded.select_dtypes(include='object'):
    df_encoded[col] = LabelEncoder().fit_transform(df_encoded[col])
plt.figure(figsize=(8, 6))
sns.heatmap(df_encoded.corr(), annot=True, cmap='coolwarm')
plt.title('Feature Correlation Heatmap (Encoded)')
plt.show()
```

**Swarm Plot**

```python
import seaborn as sns
import matplotlib.pyplot as plt
features = ['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']
for feature in features:
    plt.figure(figsize=(7, 4))
    sns.swarmplot(x='Species', y=feature, data=df)
    plt.title(f'{feature} - Swarm Plot by Species')
    plt.show()
```

**Histogram**

```python
df['PetalLengthCm'].hist(bins=20)
plt.title('Distribution of Petal Length')
plt.xlabel('Petal Length (cm)')
plt.ylabel('Frequency')
plt.show()
```

**Radar Chart (Spider Plot)**

```python
import numpy as np

# Prepare data
mean_values = df.groupby('Species').mean()
labels = mean_values.columns
species = mean_values.index
angles = np.linspace(0, 2 * np.pi, len(labels), endpoint=False).tolist()
angles += angles[:1]

# Plot each species
plt.figure(figsize=(6, 6))
for sp in species:
    values = mean_values.loc[sp].tolist()
    values += values[:1]  # repeat first value
    plt.polar(angles, values, label=sp)

plt.xticks(angles[:-1], labels)
plt.title('Radar Chart of Mean Feature Values')
plt.legend(loc='upper right')
plt.show()
```

**Facet Grid (Multi-plot per Feature or Species)**

```python
grid = sns.FacetGrid(df, col="Species")
grid.map(sns.histplot, "PetalLengthCm", bins=10)
plt.suptitle("Petal Length Distribution by Species", y=1.1)
plt.show()
```

```python
missing = data.isnull().mean().sort_values(ascending=False)
missing = (missing * 100).round(2)
missing[missing > 0]
```

**Basic cleaning (rename, strip whitespace, drop exact duplicates)**

```python
data.columns = [c.strip() for c in data.columns]          # strip whitespace in column names
for c in data.select_dtypes(include=['object']).columns:
    data[c] = data[c].astype(str).str.strip().replace('nan', np.nan)  # normalize strings

# drop fully duplicated rows (rare)
data = data.drop_duplicates()
print("After dedupe:", data.shape)
```

```python
missing.isnull().sum()
```

**Exploratory Data Analysis (visualize key relationships)**

**BarPlot**

```python
# Survival rates by categorical features
cat_cols = ['Sex', 'Pclass', 'Embarked', 'Title', 'IsAlone', 'Deck']
for c in cat_cols:
    if c in data.columns:
        display(data.groupby(c)['Survived'].mean().sort_values(ascending=False))
        plt.figure(figsize=(6,3))
        sns.barplot(x=c, y='Survived', data=data, order=sorted(data[c].dropna().unique()))
        plt.title(f"Survival rate by {c}")
        plt.show()
```

**Heat Map**

```python
# Numeric distributions / scatter
numeric = data.select_dtypes(include=[np.number]).columns.tolist()
print("Numeric columns:", numeric)
# correlation heatmap
plt.figure(figsize=(8,6))
sns.heatmap(data[numeric].corr(), annot=True, fmt=".2f", cmap="coolwarm")
plt.title("Correlation (numeric)")
plt.show()
```

**Scatter Plot**

```python
# interactive scatter: Age vs Fare colored by Survived
import plotly.express as px 
if set(['Age', 'Fare', 'Survived']).issubset(data.columns):
    fig = px.scatter(
        data, 
        x='Age', 
        y='Fare', 
        color='Survived', 
        hover_data=['Name', 'Pclass']
    )
    fig.show()
```

**Histogram**

```python
col = "Age"  # the numeric column you want to plot
plt.figure(figsize=(8,4))
plt.hist(data[col].dropna(), bins=30)  # use only the numeric column, drop NaN values
plt.xlabel(col)
plt.ylabel("Count")
plt.title(f"Distribution of {col}")
plt.show()
```

**BoxPlot**

```python
plt.figure(figsize=(8,5))
sns.boxplot(x='Pclass', y='Age', data=data)
plt.title("Age distribution by Passenger Class")
plt.show()
```

```python
fig = px.box(data, x="Pclass", y="Age", color="Survived", points="all",
             title="Boxplot: Age by Class & Survival")
fig.show()
```
